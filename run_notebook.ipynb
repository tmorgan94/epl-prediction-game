```python
# Import libraries and modules
import sys
sys.path.append('./src')
from data_loader import fetch_fpl_data
from data_transform import get_league_history, validate_team_names, calculate_leaderboard
```

```python
# Load Data
df_predictions, df_teams, df_fixtures = fetch_fpl_data()
```

```python
# Transform Data
validate_team_names(
    df_predictions=df_predictions,
    df_teams=df_teams)

df_league = get_league_history(
    df_fixtures=df_fixtures, 
    df_teams=df_teams)

df_leaderboard, df_merged = calculate_leaderboard(
    df_predictions=df_predictions, 
    df_league=df_league
)
```

```python
latest_gw = df_leaderboard['gameweek'].max()
df_leaderboard_latest = (
    df_leaderboard[df_leaderboard['gameweek'] == latest_gw]
    .sort_values('total_score', ascending=False)
    .reset_index(drop=True)
)
df_leaderboard_latest.head(7)
```
